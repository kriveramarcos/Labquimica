Usando PeriodicTable, crea una tabla con 20 elementos químicos y en las columnas considera: nombre, simbolo, masa_atomica, numero_atomico	y valencias. Puedes ayudarte del siguiente [material](https://github.com/inefable12/UNMSM_quimica_2023/blob/main/Python_para_quimica_2023.ipynb).

```python
from rdkit import Chem

PeriodicTable = Chem.GetPeriodicTable()
```

```python
help(PeriodicTable)
```

```python
#Gracias a la herramienta help encontramos que "GetNOuterElecs" nos permite encontrar los electrones de valencia
```

```python
# Generamos una lista de 20 elementos como se pide

element_symbols = []

for i in range(1,21):
  element_symbols.append(PeriodicTable.GetElementSymbol(i))
```

```python
#Imprimimos la lista generada
element_symbols
```

```python
#Notamos que con la herramienta PeriodicTable podemos ingresar datos de los elementos químicos y nos
# mostraran resultados como número atómico, número de masa, etc (se generó el número atómico a manera de ejemplo)
PeriodicTable.GetAtomicNumber('Ne')
```

```python
# Guardamos las informaciones pedidas (nombre, simbolo, masa_atomica, numero_atomico y valencias) en listas
numero_atomico = [PeriodicTable.GetAtomicNumber(x) for x in element_symbols]
masa_atomica = [PeriodicTable.GetAtomicWeight(x) for x in element_symbols]
nombre_del_elemento = [PeriodicTable.GetElementName(x) for x in range(1,21)]
electrones_valencia = [PeriodicTable.GetNOuterElecs(x) for x in range(1,21)]
```

```python
# importamos PandasTools de la librería rdkit.Chem
from rdkit.Chem import PandasTools
import pandas as pd
import os
from rdkit import RDConfig
```

```python
#Añadimos los valores que queremos representar para que esten vinculados con la variable data y así poder mostar los resultados
data = {'nombre': nombre_del_elemento, 'simbolo': element_symbols,
        'masa_atomica': masa_atomica, 'numero_atomico': numero_atomico,
        'electrones_valencia': electrones_valencia}
df = pd.DataFrame(data)
```

```python
#La reacción química que obtenemos es la siguiente
print(' C4H8 +  O2 ->  CO2 +  H2O')
```

```python
#Vamos a crear un pequeño programa para representar la reacción en formato unicode
def convertir_a_subindice(cadena):
    subindice_map = {
        '0': '₀', '1': '₁', '2': '₂', '3': '₃', '4': '₄',
        '5': '₅', '6': '₆', '7': '₇', '8': '₈', '9': '₉'
    }
    subindice_cadena = ''.join(subindice_map[digito] if digito in subindice_map else digito for digito in cadena)
    return subindice_cadena.replace('->', '→')

# Usaremos este programa para convertir nuestra reacción a formato unicode
reaccion = 'C4H8 +  O2 ->  CO2 +  H2O'
reaccion_subindice = convertir_a_subindice(reaccion)

print(f'Reacción: {reaccion}')
print(f'Reacción con unicode: {reaccion_subindice}')
```

```python
# Balanceando la reacción y asignando los subíndices se obtiene
print('La reacción química es: C₄H₈ +  O₂ →  CO₂ +  H₂O')
print('Balanceando se obtiene: 1C₄H₈ + 6 O₂ → 4 CO₂ + 4 H₂O')
```
