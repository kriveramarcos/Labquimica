```python
df = pd.read_csv("data.csv")
df
```

```python
#Realizamos la regresión lineal
plt.scatter(1/df["x"],df["y"])
```

```python
# Utilizamos la biblioteca scikit-learn para realizar la regresión lineal
from sklearn.linear_model import LinearRegression

# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo a tus datos
modelo.fit(df[['x']], df['y'])
```

```python
#Obtenemos los coeficientes de la ecuacion
intercepto = modelo.intercept_
pendiente = modelo.coef_[0]
```

```python
#Generamos la ecuacion
ecuacion = f'x = {pendiente:.2f}y + {intercepto:.2f}'
print("La ecuación será:",ecuacion)
```
